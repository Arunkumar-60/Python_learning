```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.arange(11)
y= x**2
print(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('x-axis Title')
plt.ylabel('Y-axis Title')
plt.title('figure/plot Title')
```

```python
# plt.subplot(nrows,ncols,plot_number)
plt.subplot(1,3,1)
# (131) is same as (1,3,1)
plt.plot(x,y,'r--')

plt.subplot(1,3,2)
plt.plot(y,x,'g*-')

plt.subplot(1,3,3)
plt.plot(x,y,'r--')
plt.plot(y,x,'g*-')
```

```python
fig1=plt.figure()

axes=fig1.add_axes([0.1,0.1,0.9,0.9]) #this defines  the size of the graph

axes.plot(x,y,'b') 

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Figure Title')
```

```python
fig2=plt.figure()
axes_main=fig2.add_axes([0.1,0.1,0.9,0.9]) #main set size , coordinates of position of corners
axes_inset=fig2.add_axes([0.3,0.5,0.3,0.4]) #inset size
#main figure
axes_main.plot(x,y,'b')
axes_main.set_xlabel('X-label_main') #x axis name
axes_main.set_ylabel('Y-label_main') #y axis name
axes_main.set_title('Title_inset')

#inset figure
axes_inset.plot(y,x,'r')
axes_inset.set_xlabel('X-label_inset') #same x axis anme but inset
axes_inset.set_ylabel('Y-label_inset')
axes_inset.set_title('Title_inset');
```
